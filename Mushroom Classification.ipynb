```python
dataset["class"] = dataset["class"].map({"e":0,"p":1})
dataset["cap-shape"] = dataset["cap-shape"].map({"b":0, "c":1, "x":2, "f":3, "k":4, "s":5})
dataset["cap-surface"] = dataset["cap-surface"].map({"f":0, "g":1, "y":2, "s":3})
dataset["cap-color"] = dataset["cap-color"].map({"n":0, "b":1, "c":2, "g":3, "r":4, "p":5, "u":6, "e":7, "w":8, "y":9})
dataset["bruises"] = dataset["bruises"].map({"t":0, "f":1})
dataset["odor"] = dataset["odor"].map({"a":0,"l":1,"c":2,"y":3,"f":4,"m":5,"n":6,"p":7,"s":8})
dataset["gill-attachment"] = dataset["gill-attachment"].map({"a":0,"d":1,"f":2,"n":3})
dataset["gill-spacing"] = dataset["gill-spacing"].map({"c":0, "w":1, "d":2})
dataset["gill-size"] = dataset["gill-size"].map({"b":0, "n":1})
dataset["gill-color"] = dataset["gill-color"].map({"k":0,"n":1,"b":2,"h":3,"g":4,"r":5,"o":6,"p":7,"u":8,"e":9,"w":10,"y":11})
dataset["stalk-shape"] = dataset["stalk-shape"].map({"e":0, "t":1})
dataset["stalk-root"] = dataset["stalk-root"].map({"b":0,"c":1,"u":2,"e":3,"z":4,"r":5,"?":6})
dataset["stalk-surface-above-ring"] = dataset["stalk-surface-above-ring"].map({"f":0,"y":1,"k":2,"s":3})
dataset["stalk-surface-below-ring"] = dataset["stalk-surface-below-ring"].map({"f":0,"y":1,"k":2,"s":3})
dataset["stalk-color-above-ring"] = dataset["stalk-color-above-ring"].map({"n":0,"b":1,"c":2,"g":3,"o":4,"p":5,"e":6,"w":7,"y":8})
dataset["stalk-color-below-ring"] = dataset["stalk-color-below-ring"].map({"n":0,"b":1,"c":2,"g":3,"o":4,"p":5,"e":6,"w":7,"y":8})
dataset["veil-type"] = dataset["veil-type"].map({"p":0, "u":1})
dataset["veil-color"] = dataset["veil-color"].map({"n":0,"o":1,"w":2,"y":3})
dataset["ring-number"] = dataset["ring-number"].map({"n":0, "o":1, "t":2})
dataset["ring-type"] = dataset["ring-type"].map({"c":0,"e":1,"f":2,"l":3,"n":4,"p":5,"s":6,"z":7})
dataset["spore-print-color"] = dataset["spore-print-color"].map({"k":0,"n":1,"b":2,"h":3,"r":4,"o":5,"u":6,"w":7,"y":8})
dataset["population"] = dataset["population"].map({"a":0,"c":1,"n":2,"s":3,"v":4,"y":5})
dataset["habitat"] = dataset["habitat"].map({"g":0,"l":1,"m":2,"p":3,"u":4,"w":5,"d":6})
```

```python
X = dataset.iloc[:, 1:]
y = dataset.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.1, max_iter=50)
lr.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

parameter = {
    "C":[0.1, 0.4, 0.5, 0.8, 1.0, 1.1],
    "max_iter": [50,100,200,500,1000]
}

grid_search = GridSearchCV(lr, param_grid=parameter, cv=10, scoring="accuracy")
```

```python
# grid_search.fit(X_train, y_train)
# grid_search.best_params_
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

models = [pac, svm, knn, lr, rf, dt]
model_name = ["PAC", "SVM", "KNN", "Log Reg", "Rand Forest", "Dec Tree"]
akurasi = []
urutan = 0

for model in models:
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_pred, y_test)
    acc = accuracy_score(y_pred, y_test)
    akurasi.append(acc)
    
#     plt.figure(figsize=(10,5))
    plt.title(f"{model_name[urutan]} Confusion Matrix: {round(acc*100,2)} % Accuracy")
    sns.heatmap(cm, cmap="Blues", annot=True, fmt="d")
    plt.show()
    
    urutan += 1
```

```python
plt.figure(figsize=(15,7))
plt.bar(model_name, akurasi)
plt.title("Akurasi Prediksi Mushroom")
plt.ylabel("Persentase")
plt.xlabel("Algorithm Classifier")
plt.ylim(0.9, 1.01)
plt.show()
```
